```python
R'''
We can change everything in this
for the 7k dataset '''

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
```

```python
#-----------------------------
# Wide Data
#-----------------------------

# Load Data X ----
Wide_X = pd.read_csv('Wide_X.csv')
# Drop Unnamed Column 
Wide_X = Wide_X.drop("Unnamed: 0", axis = 1)

# Normalize from 0.01 to 1
data_wide = Wide_X
scaler = MinMaxScaler()
scaler.fit(data_wide)

print("length: " + str(len(data_wide)))
data_wide.head()
```

```python
# Load Data Y ----
Y = pd.read_csv('Y.csv')
# Drop Unnamed Column 
Y = Y.drop("Unnamed: 0", axis = 1)
Y.head(5)
```

```python
#-------------------------
# Turn to Nested Format 
# ------------------------
data_nested = data_wide.apply(lambda inten: inten.tolist(), axis = 1).reset_index()
data_nested = data_nested.rename(columns={0: "Activity"})
data_nested.head()
```

```python
#-----------
# Smooth out data (Savitzky-Golay filter) 

from scipy.signal import savgol_filter
data_nested["Activity"] = data_nested["Activity"].apply(lambda act: savgol_filter(act, 51, 3))
data_nested.head()
```
